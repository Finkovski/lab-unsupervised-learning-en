# Challenge 2 - Data Cleaning and Transformation

If your conclusion from the previous challenge is the data need cleaning/transformation, do it in the cells below. However, if your conclusion is the data need not be cleaned or transformed, feel free to skip this challenge. But if you do choose the latter, please provide rationale.

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Drop the helper column if previously added
df = pd.read_csv("Wholesale customers data.csv")  # reload to start clean

# --- Step 1: Convert categorical columns ---
df['Channel'] = df['Channel'].astype('category')
df['Region'] = df['Region'].astype('category')

# --- Step 2: Remove extreme outliers using IQR ---
numerical_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1
df_clean = df[~((df[numerical_cols] < (Q1 - 1.5 * IQR)) | (df[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

# --- Step 3: Log-transform skewed numerical features ---
df_transformed = df_clean.copy()
df_transformed[numerical_cols] = df_transformed[numerical_cols].apply(lambda x: np.log1p(x))

# --- Step 4: Standardize for later modeling ---
scaler = StandardScaler()
df_transformed[numerical_cols] = scaler.fit_transform(df_transformed[numerical_cols])

# Preview cleaned and transformed dataset
df_transformed.head()
```

**Your comment here**

- Categorical columns (Channel, Region) were successfully converted to categorical data types.

- Outlier removal using the IQR method effectively reduced extreme values, improving data quality for modeling.

- Log transformation was applied to reduce skewness in features like Fresh, Milk, and Grocery, making the data more normally distributed.

- Standardization scaled all numerical features to have zero mean and unit variance, which is essential for distance-based algorithms like KMeans or PCA.

- The dataset is now cleaned, normalized, and ready for clustering, dimensionality reduction, or other machine learning tasks.

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:

from sklearn.preprocessing import StandardScaler

# Reload the original dataset to avoid contamination from earlier steps
df = pd.read_csv("Wholesale customers data.csv")

# Optional: Convert categorical columns
df['Channel'] = df['Channel'].astype('category')
df['Region'] = df['Region'].astype('category')

# Select only numerical features for scaling
numerical_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

# Log-transform to reduce skewness
df_log = df[numerical_cols].apply(lambda x: np.log1p(x))

# Scale the log-transformed data
scaler = StandardScaler()
customers_scale = scaler.fit_transform(df_log)

# If you want to keep it in a DataFrame with the original column names
customers_scale = pd.DataFrame(customers_scale, columns=numerical_cols)

# Preview
customers_scale.head()
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("Wholesale customers data.csv")

# Convert categorical columns
df['Channel'] = df['Channel'].astype('category')
df['Region'] = df['Region'].astype('category')

# Select and transform numerical features
numerical_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
df_log = df[numerical_cols].apply(lambda x: np.log1p(x))
scaler = StandardScaler()
customers_scale = scaler.fit_transform(df_log)

# --- KMeans Clustering ---
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['labels_KMeans'] = kmeans.fit_predict(customers_scale)

# --- DBSCAN Clustering ---
dbscan = DBSCAN(eps=0.5)
df['labels_DBSCAN'] = dbscan.fit_predict(customers_scale)

# Preview the dataset with both labels
df[['labels_KMeans', 'labels_DBSCAN']].value_counts()
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Try multiple cluster values
cluster_range = [2, 3, 4, 5, 6]

# Feature pair to visualize
x_col = 'Detergents_Paper'
y_col = 'Milk'

# Set up plot grid
fig, axes = plt.subplots(nrows=1, ncols=len(cluster_range), figsize=(20, 4))
fig.suptitle(f"KMeans Clustering: {x_col} vs {y_col} for Different Cluster Counts", fontsize=16)

# Run KMeans for each k and plot
for i, k in enumerate(cluster_range):
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(customers_scale)
    
    sns.scatterplot(
        ax=axes[i],
        x=df[x_col],
        y=df[y_col],
        hue=cluster_labels,
        palette='viridis',
        legend=False
    )
    axes[i].set_title(f"{k} Clusters")
    axes[i].set_xlabel(x_col)
    axes[i].set_ylabel(y_col)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

**Your comment here**

After experimenting with different values of `n_clusters`, we visually analyzed the cluster separation and structure using the `Detergents_Paper` vs `Milk` feature pair.

#### Observations:
- **2 Clusters**: Simple and clear split, but may oversimplify customer groups.
- **3 Clusters**: Well-balanced separation with distinct patterns; captures major variations.
- **4-6 Clusters**: Increasing fragmentation; some clusters appear very close or overlapping, suggesting possible overfitting.

###  Conclusion:
**3 clusters** appears to be the optimal choice for this dataset.  
It achieves a good balance between **simplicity and meaningful separation**, making it suitable for understanding and targeting different customer segments.


# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN

# Define combinations of eps and min_samples to try
param_grid = [
    (0.3, 5),
    (0.5, 5),
    (0.7, 5),
    (0.5, 3),
    (0.5, 10)
]

# Feature pair to visualize
x_col = 'Detergents_Paper'
y_col = 'Milk'

# Set up subplot grid
fig, axes = plt.subplots(1, len(param_grid), figsize=(20, 4))
fig.suptitle(f"DBSCAN: {x_col} vs {y_col} with Varying eps and min_samples", fontsize=16)

# Apply DBSCAN for each parameter combination and plot
for i, (eps_val, min_samples_val) in enumerate(param_grid):
    db = DBSCAN(eps=eps_val, min_samples=min_samples_val)
    labels = db.fit_predict(customers_scale)
    
    sns.scatterplot(
        ax=axes[i],
        x=df[x_col],
        y=df[y_col],
        hue=labels,
        palette='viridis',
        legend=False
    )
    axes[i].set_title(f"eps={eps_val}, min_samples={min_samples_val}")
    axes[i].set_xlabel(x_col)
    axes[i].set_ylabel(y_col)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

**Your comment here**

We experimented with different combinations of `eps` and `min_samples` to evaluate how DBSCAN clustering changes. The feature pair used for visualization was **`Detergents_Paper` vs `Milk`**.

#### Observations:

- **eps = 0.3, min_samples = 5**:
  - Many points labeled as noise (`-1`)
  - Small, tight clusters; very sensitive to density

- **eps = 0.5, min_samples = 5**:
  - Balanced clustering
  - Reduced noise, some clear group structure observed

- **eps = 0.7, min_samples = 5**:
  - Fewer clusters
  - Clusters become too broad, some overlap visible

- **eps = 0.5, min_samples = 3**:
  - Detects more clusters
  - Potential over-fragmentation and less meaningful separation

- **eps = 0.5, min_samples = 10**:
  - More conservative clustering
  - Fewer clusters formed, more points labeled as noise

###  Conclusion:
The combination of **`eps = 0.5` and `min_samples = 5`** provided the best balance between identifying meaningful clusters and minimizing noise points.  
This setting allows DBSCAN to effectively distinguish dense regions without over- or under-clustering the dataset.
